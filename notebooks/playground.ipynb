```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import glob
```

```python
data_path = r'/media/avi/54561652561635681/datasets/ultrasound-kaggle/Dataset_BUSI_with_GT/benign'
fnames = glob.glob(data_path + '/*.png')
fnames_mask = [f for f in fnames if f.split('/')[-1].split('_')[-1][:-4] == 'mask']
fnames = list(set(fnames) ^ set(fnames_mask))
fnames = sorted(fnames)
fnames_mask = sorted(fnames_mask)
fnames = sorted([f.replace('_mask', '') for f in fnames_mask if f.replace('_mask', '') in fnames])
print(len(fnames), len(fnames_mask))
```

```python
i = 100
img = Image.open(fnames[i])
mask = Image.open(fnames_mask[i])
fig, ax = plt.subplots(nrows=2, figsize=(16, 16))
ax[0].imshow(img)
ax[1].imshow(mask)
plt.show()
```
